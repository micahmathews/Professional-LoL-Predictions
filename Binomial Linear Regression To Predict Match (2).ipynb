In this Jupyter Notebook we are looking to determine how significant champion patch strength and player champion strength are for determining the outcome of a game in professional League of Legends

```python
import pandas as pd

# create a list of file names
file_names = ['lol2015.csv', 'lol2016.csv', 'lol2017.csv', 'lol2018.csv', 'lol2019.csv', 'lol2020.csv', 'lol2021.csv', 'lol2022.csv']

# create an empty list to store the dataframes
dfs = []

# loop through the file names then append to the list
for file_name in file_names:
    df = pd.read_csv(file_name)
    dfs.append(df)

# concatenate all dataframes into one
df = pd.concat(dfs)

# reset the index 
df.reset_index(drop=True, inplace=True)

df
```

```python
#set the visible columns to the maximum determine which columns to keep
pd.set_option('display.max_columns', None)

df
```

```python
#choose the columns you want
df = df[["gameid","patch","playername", "side","position", "teamname", "champion", "result"]]
df = df.dropna(axis=0)
df
```

```python
# calculate player win rates
df['player_winrate'] = df.groupby(['playername', 'champion'])['result'].transform('mean')

# calculate champion win rates for each patch
champion_winrate_by_patch = df.groupby(['patch', 'champion'])['result'].mean().reset_index()
champion_winrate_by_patch = champion_winrate_by_patch.rename(columns={'result': 'champion_winrate_by_patch'})

# merge the patch win rates
df = df.merge(champion_winrate_by_patch, on=['patch', 'champion'])
```

```python
# calculate team average winrate
df['team_average_winrate'] = df.groupby(['gameid', 'teamname'])['player_winrate'].transform('mean')

# calculate champion average winrate by patch
df['champion_winrate_by_patch_avg'] = df.groupby(['gameid', 'teamname', 'patch'])['champion_winrate_by_patch'].transform('mean')
df
```

```python
# calculate the mean between the two new columns
df['predicted_winrate'] = (df['team_average_winrate'] + df['champion_winrate_by_patch_avg']) / 2
df
```

```python
from sklearn.linear_model import LogisticRegression

# define X and y
X = df['predicted_winrate'].values.reshape(-1, 1)
y = df['result']

# train the model
model = LogisticRegression()
model.fit(X, y)

# use the model to predict the result
predicted_result = model.predict(X)

# evaluate the model's accuracy
accuracy = model.score(X, y)
print(accuracy)
```

```python
from sklearn.model_selection import GridSearchCV

# define X and y
X = df[['team_average_winrate', 'champion_winrate_by_patch_avg']]
y = df['result']

# define the parameter grid to search
param_grid = {'C': [1.325, 1.3, 1.275]}

# call the logistic regression model
model = LogisticRegression()

# call the grid search object
grid = GridSearchCV(model, param_grid=param_grid, cv=5)

# fit the grid search object to the data
grid.fit(X, y)

# print the best parameters and the best score
print("Best parameters:", grid.best_params_)
print("Best score:", grid.best_score_)

```

```python
df["opposite_side"] = df["side"].apply(lambda x: "Red" if x == "Blue" else "Blue")


winrate_dict = {}
for idx, row in df.iterrows():
    game_id = row["gameid"]
    winrate = row["predicted_winrate"]
    winrate_dict[game_id] = winrate


df["opposite_predicted_winrate"] = df.apply(lambda x: winrate_dict[x["gameid"].replace(x["side"], x["opposite_side"])], axis=1)


df["predicted_winrate_diff"] = df["predicted_winrate"] - df["opposite_predicted_winrate"]
```

```python
num_unique_gameids = df['gameid'].nunique()
print("Number of unique game ids:", num_unique_gameids)
```

```python
#remove all rows with "predicted_winrate_diff" 0
df = df[df["predicted_winrate_diff"] != 0]
df
```

```python
#make sure we have near the same amount of unique games
num_unique_gameids = df['gameid'].nunique()
print("Number of unique game ids:", num_unique_gameids)
```

```python
from sklearn.linear_model import LogisticRegression

# define X and y
X = df['predicted_winrate_diff'].values.reshape(-1, 1)
y = df['result']

# train the model
model = LogisticRegression()
model.fit(X, y)

# use the model to predict the result
predicted_result = model.predict(X)

# evaluate the model's accuracy
accuracy = model.score(X, y)
print(accuracy)
```

```python
# define the interval size
bin_size = 0.05

# create a new column with the bin values
df['bin'] = pd.cut(df['predicted_winrate_diff'], 
                   bins=pd.interval_range(start=-1, end=1, freq=bin_size))

# group the data by the bin column and calculate the mean accuracy and count of games
accuracy_and_count_by_bin = df.groupby('bin')['result'].agg(['mean', 'count'])

# print the accuracy and count by bin table
print(accuracy_and_count_by_bin)

```
